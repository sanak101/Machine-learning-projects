```python
df1 = pd.read_csv('hiring.csv')
df1
```

```python
df1.isnull().sum()
```

```python
word_to_num = {'zero':0, 'one': 1, 'two': 2, 'three': 3,
    'four': 4, 'five': 5, 'six': 6, 'seven': 7,
    'eight': 8, 'nine': 9, 'ten': 10, 'eleven': 11 }
df1['experience'] = df1['experience'].replace('NaN',np.nan)
df1['experience'] = df1['experience'].map(word_to_num)
df1
```

```python
# Fill missing values with mean (or median)
df1['experience'] = df1['experience'].fillna(df1['experience'].mean())
df1['test_score(out of 10)'] = df1['test_score(out of 10)'].fillna(df1['test_score(out of 10)'].mean())
df1['interview_score(out of 10)'] = df1['interview_score(out of 10)'].fillna(df1['interview_score(out of 10)'].mean())
df1
```

```python
df1 = df1.rename(columns={'experience': 'exp','test_score(out of 10)':'test_score','interview_score(out of 10)':'interview_score','salary($)':'salary'})
```

```python
df1
```

```python
new_df = df1.drop('salary',axis=1)
```

```python
salary = df1.salary
salary
```

```python
reg1 = linear_model.LinearRegression()
reg1.fit(new_df,salary)
```

```python
reg1.predict([[2, 9, 6]])
```

```python
reg1.predict([[12, 10, 10]])
```

```python
d = pd.read_csv('hiring.csv')
d
```

```python
d.experience = d.experience.fillna('zero')
d
```

```python
d.experience = d.experience.apply(w2n.word_to_num)
d
```

```python
import math
median_test_score = math.floor(d['test_score(out of 10)'].mean())
median_test_score
```

```python
d['test_score(out of 10)'] = d['test_score(out of 10)'].fillna(median_test_score)
d
```

```python
reg = linear_model.LinearRegression()
reg.fit(d[['experience','test_score(out of 10)','interview_score(out of 10)']],d['salary($)'])
```

```python
reg.predict([[2,9,6]])
```

```python
reg.predict([[12,10,10]])
```
