```python
import pandas as pd
from sklearn import linear_model
```

```python
df = pd.read_csv('homeprices.csv')
df
```

```python
dummies = pd.get_dummies(df.town)
dummies
```

```python
merged = pd.concat([df,dummies],axis=1)
merged
```

```python
final = merged.drop(['town'],axis=1)
final
```

```python
final = final.drop(['west windsor'],axis=1)
final
```

```python
X = final.drop('price',axis=1)
X
```

```python
y = final.price
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
model.predict([[3400,0,0]])
```

```python
d = pd.read_csv("carprices.csv")
d
```

```python
dummies = pd.get_dummies(d['Car Model'])
dummies
```

```python
merged = pd.concat([d,dummies],axis=1)
merged
```

```python
final = merged.drop(['Car Model'],axis=1)
final
```

```python
final = final.drop(['Mercedez Benz C class'],axis=1)
final
```

```python
X = final.drop(['Sell Price($)'],axis=1)
X
```

```python
y = final['Sell Price($)']
y
```

```python
model1 = LinearRegression()
```

```python
model1.fit(X,y)
```

```python
model1.coef_
```

```python
model1.predict([[59000,5,0,0]])
```

```python
score = model1.score(X,y)
score
```
